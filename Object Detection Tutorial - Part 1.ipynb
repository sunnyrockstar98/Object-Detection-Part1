#### But how do we make a computer output something like that with just a simple image or video-feed??

Well we need to have a look at what is called image classification and object detection.

There is a black-box here - 

**Image Classification** is taking an image, feeding it into a network of neurons( a special network called **CNNs**) and getting the image classified into one of the classes we had trained our network to predict.

**Object Detection** is a more advanced version of the Image Classification problem where it contrasts with the fact that an Object Detector will have to search an image for commonly-known( or trained) objects and if found state what object it is along with confidence of the prediction.

So if you train a CNN to predict just a lion or a tiger or a zebra that is an image classifier.

But if your program can automatically locate the animal in any image you provide and tell you - that here is the animal in this image and I am 90% confident about it!! - then that is an Object Detetctor for you

## Let's dig deeper....


#### Understanding Image Classification:

A simple convolutional neural net - looks like this:


![alt conv](conv.png)

The underlying principle is that any image can be represented as a matrix of pixel values - like every RGB image is a pixel matrix of the order (m,n,3).


A CNN takes in an image as an input applies various filters on it and various other image processing techniques so as to get a suitable matrix of numbers which can classify that image into any one of the desired number of categories we want.

#### But how can the CNN network actually know how to classify??

Well first ofcourse the network architecture has to be decided upon and then it has to be trained and only then can the network on any new image perform classification.

It is just like teaching a child what is a bird or a dog and then asking it to tell.

So the network and its neurons are initialized with random weights and trained with various images and the weights keep updating with more training.

Now when this trained network is given any image it goes through the whole process again with those weights and outputs a vector which is interpreted as a category and classification process is over.

The ConvNet has four main operations:

1. Convolution
2. Non Linearity (ReLU)
3. Pooling 
4. Classification (Fully Connected Layer)

### 1. Convolution:

For sure the electrical engineers know very well what a convolution operation is mathematically. Without touching the math a lot and intuitively put in case of  ConvNets is - 

Extract features from the input image.

Now what are features??

    - Vertica Edges in a Image
    - Horizontal Edges in a Image
    - Bright/Dark areas
    - If a person is smiling or not
    and many more.

### 2. Non-Linearity (ReLU layer):

It is basically used to convert negative pixel values into zero values.
This takes care of the non - linearity from the images as convolution was a linear operation.

So if you give ReLU layer an input feature map:


![alt relupng](relupng.png)

### 3. Pooling:

After convolution and ReLU has been applied next step in our image processing is Pooling which helps reducing unwanted dimensions from our feature map.

There are various kinds of pooling like max,average,min and many more.

These image will make it clear what happens when you apply pooling to a pixel matrix after convolution and relu:

![alt pool](pool.jpg)

You can decide on stride of window, how many pooling layers you want for your custom application.

### 4. Fully Connected Layer and Classification:

This is the final step for the CNN and our image is almost ready to be classified.

In this last output step of the CNN FC layer computes the scores for the classes which we need to classify into.
